```python
#import all libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data arrays
x = np.array([10, 100, 1000, 5000, 10000, 20000, 30000, 40000])
y1 = np.array([0 ,0 ,0.01 ,0.21 ,0.81 ,3.24, 7.31, 12.99])
y2 = np.array([0.0037, 0.001, 0.016, 0.11, 0.3673, 0.894, 1.774, 2.9407])
```

```python
plt.figure(figsize=(12, 9))
plt.plot(x, y1, linewidth = 2, label = 'Последовательный', marker = 'o', markersize = 5)
plt.plot(x, y2, linewidth = 2, label = 'Параллельный', marker = 'o', markersize = 5)
plt.legend(fontsize = 16)
plt.grid(True)
plt.title("Сравнение времени выполнения алгоритма", fontsize = 16)
plt.xlabel("Количество вершин", fontsize = 14)
plt.ylabel("Время (в секундах)", fontsize = 14)
plt.show()
```
